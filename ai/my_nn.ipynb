```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report, confusion_matrix
```

```python
transactions = pd.read_csv('FoodUtilities.csv')
# print(transactions.to_string())
transactions
```

```python
transactions.describe().transpose()
```

```python
X = transactions.drop('Categories', axis=1)
y = transactions['Categories']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
print(len(X_train), len(X_test))
```

```python
scaler = StandardScaler()
```

```python
# Fit only to the training data
scaler.fit(X_train)
```

```python
StandardScaler(copy=True, with_mean=True, with_std=True)
```

```python
# Now apply the transformations to the data:
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(5,5,5),max_iter=500)
```

```python
mlp.fit(X_train, y_train)
```

```python
predictions = mlp.predict(X_test)
predictions
```

```python
y_test
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test,predictions))
```
