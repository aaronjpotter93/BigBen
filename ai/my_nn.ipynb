```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.neural_network import MLPClassifier
from sklearn.neural_network import MLPRegressor
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import json
```

```python
# Read csv file into pandas dataframe
transactions = pd.read_excel('FoodUtilitiesGroceriesGas.xlsx')
transactions
```

```python
from dataFrameBuilder import buildDataFrame
# pd.set_option("display.max_rows", None, "display.max_columns", None)

df = buildDataFrame()
df = df.drop('Merchant Name', axis=1)
df
```

```python
transactions.describe().transpose()
```

```python
transactions = df
```

```python
X = transactions.drop('Categories', axis=1)
y = transactions['Categories']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y)
print(len(X_train), len(X_test))
```

```python
scaler = StandardScaler()
```

```python
# Fit only to the training data
scaler.fit(X_train)
```

```python
StandardScaler(copy=True, with_mean=True, with_std=True)
```

```python
# Now apply the transformations to the data:
X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
mlp = MLPClassifier(hidden_layer_sizes=(60,90,120,90,60),max_iter=100000)
```

```python
mlp.fit(X_train, y_train)
```

```python
predictions = mlp.predict(X_test)
```

```python
print(confusion_matrix(y_test, predictions))
```

```python
print(classification_report(y_test,predictions))
```

```python
# Read excel file into pandas dataframe
small_batch = pd.read_excel('small_batch.xlsx')
small_batch = transactions
small_batch
```

```python
# Separate Label from transaction
X = small_batch.drop('Categories', axis=1)
y = small_batch['Categories']
```

```python
# Scale the data
scaler = StandardScaler()
# Fit only to the training data
scaler.fit(X)
StandardScaler(copy=True, with_mean=True, with_std=True)
# Now apply the transformations to the data:
X = scaler.transform(X)
```

```python
predictions = mlp.predict(X)
predictions
```

```python
print(confusion_matrix(y, predictions))
```

```python
for (predict, actual, score) in zip(predictions, y, mlp.predict_proba(X)):
    print("prediction: ", predict, "actual: ", actual, "confidence: ", score)
```

```python
for (i, predict, actual, score) in zip(y.index, predictions, y, mlp.predict_proba(X)):
    print("DataPoint: ", i, "prediction: ", predict, "actual: ", actual, "confidence: ", score)

array = mlp.predict_proba(X).tolist()


def topThreeIndices(array):
    topThree = []
    second_array = []
    # Make deep copy of array
    for i in range(len(array)):
        second_array.append(0)
        second_array[i] = array[i]
    for i in range(3):
        max_value = max(second_array)
        max_index = array.index(max_value)
        topThree.append(max_index)
        second_array.remove(max_value)
    return topThree

results = []
for predictions in array:
    result = topThreeIndices(predictions)
    results.append(result)

with open('NeuralNetworkPredictions.json', 'w') as jsonfile:
    json_results = {}
    json_results["Predictions"] = results
    json.dump(json_results, jsonfile)
```
